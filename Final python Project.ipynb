```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('owid-covid-data.csv')
df.columns   #displaying all columns
```

```python
df.head(10)    #Printing first 10 rows
```

```python
df.isnull().sum()   #Displaying number of null values in each column
```

```python
#Filtered to three countries and dropped unnecessary columns and rows with null values 
df_filtered = df[df['location'].isin(["Kenya","India","United States"])].drop(df.columns[35:], axis=1).drop(df.columns[9:34],axis=1).dropna()
df_filtered
```

```python
#Checking Data type of columns
display(df_filtered.dtypes)
df_filtered['date'] = pd.to_datetime(df['date'])   #changed date column to date columnn instead of object
display(df_filtered)
```

```python
#Handle missing numeric values
df_filtered.fillna(0)
```

```python
kenya_data = df_filtered[df_filtered['location'] == 'Kenya']

plt.figure(figsize=(10, 5))
plt.plot(kenya_data['date'], kenya_data['total_cases'], color='green')
plt.title('Total COVID-19 Cases in Kenya')
plt.xlabel('Date')
plt.ylabel('Total Cases')
plt.tight_layout()
plt.show()

india_data = df_filtered[df_filtered['location'] == 'India']

plt.figure(figsize=(12, 6))
plt.plot(india_data['date'], india_data['total_cases'], color='green')
plt.title('Total COVID-19 Cases in India')
plt.xlabel('Date')
plt.ylabel('Total Cases')
plt.tight_layout()
plt.show()

united_states_data = df_filtered[df_filtered['location'] == 'United States']

plt.figure(figsize=(12, 6))
plt.plot(united_states_data['date'], united_states_data['total_cases'], color='green')
plt.title('Total COVID-19 Cases in U.S')
plt.xlabel('Date')
plt.ylabel('Total Cases')
plt.tight_layout()
plt.show()

plt.figure(figsize=(12, 6))

for country in df_filtered['location'].unique():
    data = df_filtered[df_filtered['location'] == country]
    plt.plot(data['date'], data['total_cases'], label=country)

plt.yscale('log')  
plt.title('Total COVID-19 Cases (Log Scale)')
plt.xlabel('Date')
plt.ylabel('Total Cases (log scale)')
plt.legend()
plt.tight_layout()
plt.show()
```

Foloowing insights are made:
1. India had the fastest rise in cases between 2021-04 to around 2021-07
2. United States had the highest recorderd cases daily
3. Compared to the others United States had a gradual increase in cases

```python
#Total death in each country over the months
plt.figure(figsize=(12, 6))

for country in df_filtered['location'].unique():
    data = df_filtered[df_filtered['location'] == country]
    plt.plot(data['date'], data['total_deaths'], label=country)

plt.yscale('log')  
plt.title('Total COVID-19 Deaths Over Time (Log Scale)')
plt.xlabel('Date')
plt.ylabel('Total Deaths (log scale)')
plt.legend()
plt.tight_layout()
plt.show()
```

In above plot for Total Covid Deaths Over Time, following insights are made:
1. Kenya has lowest total deaths
2. United States with highest total deaths

```python
#Comparison of daily cases in each country
plt.figure(figsize=(12, 6))

for country in df_filtered['location'].unique():
    data = df_filtered[df_filtered['location'] == country]
    plt.plot(data['date'], data['new_cases'], label=country)

plt.yscale('log')  
plt.title('Daily New COVID-19 Cases (Log Scale)')
plt.xlabel('Date')
plt.ylabel('New Cases (log scale)')
plt.legend()
plt.tight_layout()
plt.show()
```

Insights:
1. India experienced two major sharp waves of new cases followed by smaller more frequent waves.
2. Kenya experienced a major spike in cases around early 2022 
3. United States had several significant waves of cases throughout the period

```python
#Death Rate Over Time
df_filtered['death_rate'] = df_filtered['total_deaths'] / df_filtered['total_cases']

plt.figure(figsize=(12, 6))

for country in df_filtered['location'].unique():
    data = df_filtered[df_filtered['location'] == country]
    plt.plot(data['date'], data['death_rate'], label=country)

plt.title('COVID-19 Death Rate Over Time')
plt.xlabel('Date')
plt.ylabel('Death Rate')
plt.legend()
plt.tight_layout()
plt.show()
```

Insights:
1. India showed a significant surge in death rate during major waves of infection
2. Kenya experienced a relatively lower death rate compared to US and India throughout the period
3. United had a substantial increase in death rate during initial waves

```python

countries = ['Kenya', 'India', 'United States']


latest = df_filtered.sort_values('date').groupby('location').tail(1)


fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(12, 4))


for i, country in enumerate(countries):
    value = latest[latest['location'] == country]['total_cases'].values[0]
    axes[i].bar(country, value, color='tomato')
    axes[i].set_title(f"{country}\nTotal Cases")
    axes[i].set_ylabel('Total Cases')

plt.tight_layout()
plt.show()
```

1. United States shows highest total number of Covid cases
2. India has second highest number of cases
3. Kenya exhibits lowest total number of cases

```python
#Plot Cumulative Vaccinations Over Time
plt.figure(figsize=(12, 6))

for country in df_filtered['location'].unique():
    data = df_filtered[df_filtered['location'] == country]
    plt.plot(data['date'], data['total_vaccinations'], label=country)

plt.title('Cumulative COVID-19 Vaccinations Over Time (Log Scale)')
plt.xlabel('Date')
plt.ylabel('Total Vaccinations')
plt.yscale('log')  
plt.legend()
plt.tight_layout()
plt.show()
```

1. United States shows highest cumulative number of vaccinations reached by early 2023
2. India achived a substantial cumulative number of vaccinations
3. Kenya had a lower cumulative number of vaccintions comapred to India and U.S throughout the period

```python
#Total COVID-19 Cases by Country
import plotly.express as px


latest_cases = df_filtered.sort_values('date').groupby('location').tail(1)

latest_cases = latest_cases.dropna(subset=['total_cases', 'iso_code'])

# Plot
fig = px.choropleth(
    latest_cases,
    locations='iso_code',
    color='total_cases',
    hover_name='location',
    color_continuous_scale='OrRd',
    title='COVID-19: Total Reported Cases by Country',
    projection='natural earth'
)

fig.show()


```
